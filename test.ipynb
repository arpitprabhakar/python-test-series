<h2>Q1</h2>

```python
def top_k(items,k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
items=[("A",90), ("B",95), ("C",95), ("D",85)]
k=2
result=top_k(items,k)
print(result)
```

<h2>Q2</h2>

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat,{})
        d[cat].setdefault(word, 0)
        d[cat][word]+=1
    return d
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
print(count_by_category(pairs))
```

<h2>Q3</h2>

```python
def missing_number(nums):
    n=len(nums)
    total=n*(n+1)/2
    return int(total-sum(nums))
nums=[3,0,1]
print(missing_number(nums))
```

<h2>Q4</h2>

```python
def add_unique(x,seen=None):
    if seen is None:
        seen=set()
    if x not in seen:
        seen.add(x)
    return list(seen)
print(add_unique(1))
print(add_unique(2))
```

<h2>Q5</h2>

```python
def moving_average(stream, w):
    result=[]
    window_sum=0
    for i in range(len(stream)):
        window_sum+=stream[i]
        if i>=w:
            window_sum-=stream[i-w]
        current_window_size=min(i+1,w)
        result.append(window_sum/current_window_size)
    return result
stream=[1,2,3,4,5]
w=3
print(moving_average(stream,w))
```
